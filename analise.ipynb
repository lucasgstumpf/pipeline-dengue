```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arquivo = 'resultadoPipeline-00000-of-00001.csv'
df = pd.read_csv(arquivo, delimiter=';')
df
```

```python
df.groupby(['UF', 'ANO'])[['CHUVA','DENGUE']].mean()
df.head()
```

```python
ufs = df['UF'].unique()
ufs
```

```python
for uf in ufs:
    uf_data = df[df['UF'] == uf]
    
    plt.figure(figsize=(10, 6))
    plt.bar(uf_data['ANO'], uf_data['CHUVA'], width=0.4, label='Chuva')
    plt.bar(uf_data['ANO'] + 0.4, uf_data['DENGUE'], width=0.4, label='Dengue')
    plt.title(f'RelaÃ§Ã£o Chuva e Dengue em {uf}')
    plt.xlabel('Ano')
    plt.ylabel('Valor')
    plt.legend()
    plt.xticks(uf_data['ANO'])
    plt.grid()
    plt.show()
```
